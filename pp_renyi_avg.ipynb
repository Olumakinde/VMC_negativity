```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sqrt
from ipywidgets import interact
import sys
from os import path
```

```python

out_dir= 'avg_files/'

ex=1
r=2 # Renyi index
# system size
N=20
Lsub_list=np.arange(1,N)
# Lsub_list=np.arange(4,10,2)
# Lsub_list=[8]
N_pt = int(N/2)

# reference slater determinant
t= -0.

f1='R%d_e_%d_N_%d_Npt_%d_t_%.2f.npz' % (r,ex,N,N_pt,t)
fname= out_dir + f1 

if path.exists(fname):
    npzfile = np.load(fname)
    Rr_vmc = npzfile['Rr_vmc']
    acc_ratio = npzfile['acc_ratio']
    counter = npzfile['counter']
    if ex==1:
        Rr_exact= npzfile['Rr_ex']
            
print('Average is over : ', counter, ' x 10k' )


```

```python
# print('vmc: ', R2_vmc)
# print('exact: ', R2_ex)
print('error [%]:', np.abs(Rr_vmc-Rr_exact)/Rr_exact*100)

plt.figure()
if ex==1:
    plt.plot(Lsub_list,Rr_exact,'o')
plt.plot(Lsub_list,Rr_vmc,'x')
plt.xlabel('Subsystem length')
plt.ylabel('r=%d Renyi entropy' % (r))
plt.show()

if ex==1:
    plt.figure()
    plt.plot(Lsub_list,np.abs(Rr_vmc-Rr_exact)/Rr_exact*100,'o')
    plt.ylabel('Error [%]')
    plt.xlabel('Subsystem length')
    plt.show()

plt.figure()
plt.plot(Lsub_list,100*acc_ratio,'o')
plt.ylabel('Acc ratio [%]')
plt.xlabel('Subsystem length')
plt.show()
```
